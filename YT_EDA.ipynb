```python
## CPTV return NaN when view_count = 0, to prevent that we need to develop the formula using conditions

# df = df.drop('cptv', axis = 1)
```

```python
df.info()
```

```python
# Use np.where to calculate only when view_count > 0

df['cptv'] = np.where(
    df['view_count']>0,
    df['comment_count']/df['view_count']*1000,
    0     ## To handle 0 view_count
)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# Engagement_rate, likes_per_1000_views

df['likes_per_1000_views'] = np.where(
    df['view_count']>0,
    df['like_count']/df['view_count']*1000,
    0
)

df['Engagement_rate'] = np.where(
    df['view_count']>0,
    (df['comment_count']+df['like_count'])/df['view_count']*1000,
    0
)
```

# Business question

## Q1. Which channels are getting the most views overall?

```python
df.info()
```

```python
# top performing creator by viewership

channel_views = (df.groupby('channel_title', as_index = False)['view_count'].sum().sort_values('view_count', ascending = False))
channel_views.head(10)  #top 10
```

```python
# top performing creator by likes

df.groupby('channel_title', as_index = False)['like_count'].sum().sort_values('like_count', ascending = False).head(10)
```

```python
# channel-level
channel_kpis = df.groupby('channel_title', as_index=False).agg(
    videos=('Title','count'),
    total_views=('view_count','sum'),
    avg_views=('view_count','mean'),
    videos_count=('Title','count'),
    engagement_mean=('Engagement_rate','mean')
).sort_values('total_views', ascending=False)

# video-level top performers
top_videos = df.sort_values('view_count', ascending=False)[[
    'Title','channel_title','view_count','like_count','comment_count','cptv'
]]
```

```python
print(channel_kpis.head(10))
print(top_videos.head(10))
```

```python
# Monthly time series
df['pub_month'] = df['published_at'].dt.to_period('M').dt.to_timestamp()
monthly = df.groupby('pub_month').agg(
    total_views=('view_count','sum'),
    videos_published=('Title','count')
).reset_index()
print(monthly)
```

```python
#Yearly time series

df['pub_year'] = df['published_at'].dt.to_period('Y').dt.to_timestamp()
yearly = df.groupby('pub_year').agg(
    total_views=('view_count','sum'),
    videos_published=('Title','count')
).reset_index()
print(yearly)
```

```python
# Month with highest videos uploaded
monthly.loc[monthly['videos_published'].idxmax()]
```

```python
# Year with highest videos uploaded
yearly.loc[yearly['videos_published'].idxmax()]
```

## Categorical analysis

```python
# HD vs NonHD

definition_kpi = df.groupby('definition').agg(
    videos=('Title','count'),
    total_channels=('channel_title','count'),
    total_views=('view_count','sum'),
    total_likes=('like_count','sum'),
    total_duration=('duration_seconds','sum'),
    mean_Engagement_rate=('Engagement_rate','mean')
).sort_values('total_views', ascending=False)

print(definition_kpi)
```
